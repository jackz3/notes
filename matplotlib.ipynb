```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
df=pd.read_csv('./loan_train.csv')
df.head()
```

```python
#Plots in matplotlib reside within a figure object, use plt.figure to create new figure
fig=plt.figure()
#Create one or more subplots using add_subplot, because you can't create blank figure
ax = fig.add_subplot(1,1,1)
ax.hist(df['ApplicantIncome'],bins = 50,color='#8cdcb4')
#df['ApplicantIncome'].hist(bins=40,label='lala')
#plt.hist(df['ApplicantIncome'], bins=50)

plt.title('ApplicantIncome distribution')
plt.xlabel('income')
plt.ylabel('count')
ax.set_xlabel('Gender')
ax.set_ylabel('Sum of Sales')
ax.set_title("Gender wise Sum of Sales")
plt.show()
```
